```python
unique_id = len(pd.unique(daily_activity["Id"]))
  
print("# of unique Id: " + str(unique_id))
```

```python
#now lokking for missing or null values in daily_activity
missing_values_count= daily_activity.isnull().sum()
```

```python
missing_values_count[:]
```

```python
daily_activity["ActivityDate"]= pd.to_datetime(daily_activity["ActivityDate"], format="%d-%m-%Y")
```

```python
daily_activity.info()
```

```python
daily_activity["ActivityDate"].head()
```

```python
daily_activity.head(4)
```

```python
#create new column
new_cols = ['Id', 'ActivityDate', 'DayOfWeek', 'TotalSteps', 'TotalDistance', 'TrackerDistance', 'LoggedActivitiesDistance', 'VeryActiveDistance', 'ModeratelyActiveDistance', 'LightActiveDistance', 'SedentaryActiveDistance', 'VeryActiveMinutes', 'FairlyActiveMinutes', 'LightlyActiveMinutes', 'SedentaryMinutes', 'TotalExerciseMinutes', 'TotalExerciseHours', 'Calories']
```

```python
df_activity= daily_activity.reindex(columns= new_cols)
```

```python
df_activity.head(4)
```

```python
# now seperate the date into day of week
df_activity["Dayofweek"]= df_activity["ActivityDate"].dt.day_name()
```

```python
df_activity["Dayofweek"].head(12)
```

```python
#column rename
df_activity.rename(columns = {"Id":"id", "ActivityDate":"date", "Dayofweek":"day_of_week", "TotalSteps":"total_steps", "TotalDistance":"total_dist", "TrackerDistance":"track_dist", "LoggedActivitiesDistance":"logged_dist", "VeryActiveDistance":"very_active_dist", "ModeratelyActiveDistance":"moderate_active_dist", "LightActiveDistance":"light_active_dist", "SedentaryActiveDistance":"sedentary_active_dist", "VeryActiveMinutes":"very_active_mins", "FairlyActiveMinutes":"fairly_active_mins", "LightlyActiveMinutes":"lightly_active_mins", "SedentaryMinutes":"sedentary_mins", "TotalExerciseMinutes":"total_mins","TotalExerciseHours":"total_hours","Calories":"calories"}, inplace = True)
```

```python
print(df_activity.columns)
```

```python
print(df_activity.columns.values)
```

```python
df_activity.head(10)
```

```python
#creating new column total_min
df_activity["total_min"]= df_activity["very_active_mins"]+df_activity["fairly_active_mins"]+df_activity["lightly_active_mins"]+df_activity["sedentary_mins"]
```

```python
df_activity["total_min"].head(7)
```

```python
# now to convert total_min into hours upto two decimal places
df_activity["total_hours"]= round(df_activity["total_min"] /60)
```

```python
df_activity["total_hours"].head(5)
```

```python
#plotting histogram
```

```python
plt.style.use("default")
plt.figure(figsize=(7,5))
plt.hist(df_activity.day_of_week, bins=7, width= 0.7, color= "lightgreen", edgecolor= "black")



plt.xlabel("Day of week")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
#plotting scatter plot
```

```python
plt.style.use("default")
plt.figure(figsize=(7,5))
plt.scatter(df_activity.TotalSteps,df_activity.Calories, alpha= 0.8, c= df_activity.Calories, cmap= "Spectral")


#adding annotations
median_calories= 2303
median_steps= 7637
plt.colorbar(orientation= "vertical")
plt.axvline(median_steps, color= "green", label= "Median steps")
plt.axhline(median_calories, color= "red", label= "Median calories burned")
plt.xlabel("Steps Taken")
plt.ylabel("Calories Burned")
plt.title("calories burned for each step taken")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# from the above scatter plot i noticed
1. #More the steps taken more the calories burned
```

```python
# now for the next scatter plot build relationship between hours and calories burn , means calories burn for every hour
```

```python

# plotting scatter plot
plt.style.use("default")
plt.figure(figsize=(7,6)) 
plt.scatter(df_activity.total_hours, df_activity.calories, 
            alpha = 0.8, c = df_activity.total_hours, cmap = "Spectral")
            

# adding annotations and visuals
median_hours = 20
median_sedentary = 991 / 60
median_calories = 2303

plt.colorbar(orientation = "vertical")
plt.axvline(median_hours, color = "green", label = "Median hours")
plt.axvline(median_sedentary, color = "blue", label = "Median sedentary")
plt.axhline(median_calories, color = "Red", label = "Median calories")
plt.xlabel("Hours")
plt.ylabel("Calories burned")
plt.title("Calories burned for every hour")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
```

```python
# calculating individual of every minutes column
very_active_mins= df_activity["very_active_mins"].sum()
fairly_active_mins	= df_activity["fairly_active_mins"].sum()
lightly_active_mins= df_activity["lightly_active_mins"].sum()
sedentary_mins= df_activity["sedentary_mins"].sum()


slices= [very_active_mins, fairly_active_mins, lightly_active_mins, sedentary_mins]
labels= ["very_active_mins", "fairly_active_mins", "lightly_active_mins","sedentary_mins"]
colours=["orange", "yellow", "lightskyblue", "red"]
explode= [0.9,0,0,0]
plt.style.use("default")
plt.pie(slices, labels= labels, colors= colours, wedgeprops= {"edgecolor":"black"}, explode= explode,autopct= "%1.1f%%")
plt.title("% of activity in minutes")
plt.tight_layout()
plt.show()
```
